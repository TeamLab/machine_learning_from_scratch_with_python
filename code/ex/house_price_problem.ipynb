## 전처리 전략

- 각각 처리하고 Column을 맞춰준다.
- 차이가 나는 변수는 아예 제외를 해준다.
- 둘을 애초에 합쳐서 하나의 전처리 기준으로 해준다.

```python
df_all = pd.concat([df_train, df_test])
```

```python
df_all.head()
```

```python
df_all.isnull().sum()
```

```python
for column_name in df_all.columns[df_all.isnull().sum() > 1000]:
    del df_all[column_name]
df_all.head()
```

```python
missing_values = {}
for column_name in df_all.columns:
    if df_all[column_name].dtypes == "object":
        mode_value = df_all[column_name].value_counts()[0]
        missing_values[column_name] = mode_value
        df_all[column_name] = df_all[column_name].fillna(mode_value)
    else:
        mean_value = df_all[column_name].mean()
        missing_values[column_name] = mean_value
        df_all[column_name] = df_all[column_name].fillna(mean_value)
```

```python
df_all.isnull().sum().sum()
```

```python
df_new_all = pd.get_dummies(df_all)
```

```python
list(df_new_all.columns)
```

```python
min_max_values = {a:[b,c] for a, b, c in zip(list(df_new_all.columns[1:]), df_new_all.max()[1:], df_new_all.min()[1:])}
min_max_values
```

```python
for column_name in min_max_values.keys():
    df_new_all[column_name] =  ( df_new_all[column_name] - min_max_values[column_name][1] ) \
        / (min_max_values[column_name][0] - min_max_values[column_name][1]) * (5 - 1) + 1
```

```python
x_train = df_new_all.iloc[:1460,1:].as_matrix()
x_test = df_new_all.iloc[1460:,1:].as_matrix()
y_train = y_data.as_matrix()
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_train, regr.predict(x_train))
mse = mean_squared_error(y_train, regr.predict(x_train))
r2 = r2_score(y_train, regr.predict(x_train))

print(mae, mse, r2)
```
