```python
import numpy as np
```

```python
X = np.load("./tatanic_X_train.npy")
y = np.load("./tatanic_y_train.npy")
```

```python
X[0]
```

```python
y[:10]
```

```python
from sklearn.ensemble import RandomForestClassifier
eclf = RandomForestClassifier(n_estimators=100, max_features=2, n_jobs=7, oob_score=True)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(eclf, X, y, cv=5).mean()
```

```python
params ={
    "n_estimators" : [10, 20, 30, 50, 100],
    "max_features" : [1,2,3,4,5,6,7, 10, 15, 20, 25, len(X[0])]
    }
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=eclf, param_grid=params, cv=5)
grid = grid.fit(X, y)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
grid.best_estimator_.oob_score_
```

```python
from sklearn.preprocessing import PolynomialFeatures

for i in range(3, 6):
    poly = PolynomialFeatures(i)
    X_poly = poly.fit_transform(X)
    eclf = RandomForestClassifier(n_estimators=30, max_features=2, n_jobs=7, oob_score=True)
    result = cross_val_score(eclf, X_poly, y, cv=5).mean()
    print(result)
```
